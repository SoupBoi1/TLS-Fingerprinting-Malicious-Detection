```python
import numpy as np
import json
import pandas as pd
import matplotlib.pyplot as plt
import requests
import os
```

```python
url = 'https://ja4db.com/api/read/'

response = requests.get(url)
data = response.json()
df_web = pd.DataFrame(data)
```

```python
dataset = 'otherset/datasets'
```

```python
daCon = pd.read_csv('otherset/desktop-apps-converted.csv')
dmCon = pd.read_csv('otherset/desktop-malware-converted.csv')
maCon = pd.read_csv('otherset/mobile-apps-converted.csv')
mmCon = pd.read_csv('otherset/mobile-malware-converted.csv')

```

```python
print(np.array(daCon.columns))
print(np.array(dmCon.columns))
print(np.array(maCon.columns))
print(np.array(maCon.columns))
print(np.array(df_web.columns))

```

```python
common_strings = list(set(np.array(daCon.columns)).intersection(np.array(dmCon.columns), np.array(maCon.columns), np.array(maCon.columns)))
```

```python
common_strings = common_strings[1:]
```

```python
attack_word = "Attack"
```

```python
final_inculsion_header = common_strings
final_inculsion_header.append(attack_word)
```

```python
final_inculsion_header
```

```python
daCon[attack_word] = 0
dmCon[attack_word] = 1
maCon[attack_word] = 0
mmCon[attack_word] = 1
```

```python
temdsf = [daCon[final_inculsion_header],dmCon[final_inculsion_header],maCon[final_inculsion_header],mmCon[final_inculsion_header]] 
dsf = pd.concat(temdsf)
```

```python
df = dsf.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df.to_csv("dataset.csv")
```

```python
label_counts = df[attack_word].value_counts()
label_percentages = df[attack_word].value_counts(normalize=True) * 100


display_distibution = pd.DataFrame(
    {"count": label_counts, "percentage": label_percentages}
)
display(display_distibution)
```
